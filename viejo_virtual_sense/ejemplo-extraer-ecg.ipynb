```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = "C:/Users/pedro/Documents/IPPG Virtual Sense/googledrive-archive/2. ORIGINAL DATA/P1LC4/P1LC4_Mobi_output.fys"
ecg_values = []
first_line_read = False

with open(file_path, 'r') as file:
    for line in file:
        
        if not first_line_read:
            first_line_read = True
            continue
        # Split the line into columns using whitespace as delimiter
        columns = line.split()

        # Extract the ECG value (third column)
        ecg_value = float(columns[2])

        # Append the ECG value to the list
        ecg_values.append(ecg_value)
```
